# Cálculo de integrales por el método de Monte Carlo

##Ejercicio 7:

**1.-  Mostrar que $\theta_1 $ es un estimador insesgado, es decir, que $E(\theta_1 ) = pc(b-a) = I$**

Para demostrar este ejercicio vamos utilizar esta ecuación: $ y = \sqrt((1-x^2))^3 $

Consiedramos aquí la estimación de integrales unidimensionales

>    $I = \int_a^b g(x)dx,$

Entonces: 

> $I = \int_0^1 \sqrt((1-x^2))$

> $I = 3\pi/16$

> $I = 0.589045 = E(\theta_1 )$





Entoces $pc(b-a) = E(\theta_1 )$

```python
import random
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
x = np.linspace(0, 2, 1000)
y = (1-x*x)**(3/2)
dentro = 0

print ()
fig = plt.figure(1, figsize=(6, 6))
for i in range(n):
    x_1 = random.uniform(0, 1)
    y_1 = random.uniform(0, 1)
    plt.scatter(x_1, y_1, s=20, c="red")
    if (1-x_1**2)**(3/2) >= y_1 :
      dentro += 1

plt.plot(x, y, color="blue", markersize=1)

plt.title("Grafica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.show()

print ()
print("El valor de pc(b-a) =" , (dentro/n))
```

Por lo tanto $E(\theta_1 ) = pc(b-a) = I$
