# Uso de los n√∫meros aleatorios



## Ejercicio 3:

Graficar las salidas obtenidas en la Tarea 1

```python
import random
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
x = np.linspace(-2, 2, 1000)
y = np.sqrt(4 - x*x) 
dentro = 0

print ()
fig = plt.figure(1, figsize=(6, 6))
for i in range(n):
    x_1 = random.uniform(-2, 2)
    y_1 = random.uniform(-2, 2)
    plt.scatter(x_1, y_1, s=20, c="red")
    if (x_1**2 + y_1**2) <= 4 :
      dentro += 1

plt.plot(x, y, color="blue", markersize=1)
plt.plot(x, -y, color="blue", markersize=1)

plt.title("Grafica de la Tarea 1")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.show()

print ()
print("El valor estimado de pi es" , 4*(dentro/n))
```

Visualizamos el comportamiento de nuestra estimaci√≥n en funci√≥n del n√∫mero de iteraciones.

```python
import random
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
print()

def f(n):
    d = 0
    p = 0
    M = []

    for i in range(n):
        x = random.uniform(-2, 2)
        y = random.uniform(-2, 2)
        if (x*x + y*y) <= 4 :
          d += 1
        p = 4*(d/(i+1))
        M.append(p)

    fig = plt.figure(1, figsize=(10, 6))
    plt.plot(M, color="green", markersize=1)
    plt.title("Grafica de la Tarea 1")
    plt.xlabel("Puntos")
    plt.ylabel("Estimaci√≥n de $\pi$")
    plt.grid()
    plt.show()

    print ()
    print("El valor estimado de pi es" , 4*(d/n))

f(n)
```
