# Cálculo de integrales por el método de Monte Carlo

##Ejercicio 8:

Estimar las integrales:


**1. $\int_0^1 (1 - x^2)^{\frac{3}{2}} dx$**

```python
import random
import numpy as np

n = int(input("Numero de puntos que desea "))
dentro = 0

for i in range(n):
    x_1 = random.uniform(0, 1)
    y_1 = random.uniform(0, 1)
   
    if (1-x_1**2)**(3/2) >= y_1 :
      dentro += 1

print ()
print("El valor de pc(b-a) =" , (dentro/n))
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
x = np.linspace(0, 2, 1000)
y = (1-x*x)**(3/2)
dentro = 0

print ()
fig = plt.figure(1, figsize=(6, 6))
for i in range(n):
    x_1 = random.uniform(0, 1)
    y_1 = random.uniform(0, 1)
    plt.scatter(x_1, y_1, s=20, c="red")
    if (1-x_1**2)**(3/2) >= y_1 :
      dentro += 1

plt.plot(x, y, color="blue", markersize=1)

plt.title("Grafica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.show()

print ()
print("El valor de pc(b-a) =" , (dentro/n))
```

**2. $\int_{-2}^2 \exp (x + x^2) dx$**


```python
import random
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
dentro = 0

for i in range(n):
    x_1 = random.uniform(-2, 2)
    y_1 = random.uniform(0, 403.4)
  
    if np.exp(x_1+x_1*x_1) >= y_1 :
      dentro += 1

print ()
print("El valor de pc(b-a) =" , 403.4*4*(dentro/n))
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
x = np.linspace(-2, 2, 90000)
y = np.exp(x+x*x)
dentro = 0

print ()
fig = plt.figure(1, figsize=(6, 6))
for i in range(n):
    x_1 = random.uniform(-2, 2)
    y_1 = random.uniform(0, 403.4)
    plt.scatter(x_1, y_1, s=20, c="red")
    if np.exp(x_1+x_1*x_1) >= y_1 :
      dentro += 1

plt.plot(x, y, color="blue", markersize=1)

plt.title("Grafica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.show()

print ()
print("El valor de pc(b-a) =" , 403.4*4*(dentro/n))
```
