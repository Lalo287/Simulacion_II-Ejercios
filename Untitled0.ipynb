```python
import random
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))

def f(n):
    d = 0
    p = 0
    M = []

    for i in range(n):
        x = random.uniform(-2, 2)
        y = random.uniform(-2, 2)
        if (x*x + y*y) <= 4 :
          d += 1
        p = 4*(d/(i+1))
        M.append(p)

    plt.plot(M, color="red", markersize=1)
    plt.xlabel("Iteraciones")
    plt.ylabel("EstimaciÃ³n")
    plt.grid()
    plt.show()

f(n)
```
