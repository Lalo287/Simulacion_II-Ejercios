# Estimación del número $\pi$




##Ejercicio 6:

Repetir el experimento, utilizando un cuarto de elipse.

```python
import random
import numpy as np
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
x = np.linspace(0, 5, 1000)
y = np.sqrt(9 - 9*(x*x/25)) 
dentro = 0

print ()
fig = plt.figure(1, figsize=(10, 6))
for i in range(n):
    x_1 = random.uniform(0, 5)
    y_1 = random.uniform(0, 3)
    plt.scatter(x_1, y_1, s=20, c="red")
    if ((x_1**2)/25 + (y_1**2)/9) <= 1 :
      dentro += 1

plt.plot(x, y, color="blue", markersize=1)

plt.title("Grafica")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.grid()
plt.show()

print ()
print("El valor estimado de pi es" , 4*(dentro/n))
```

Visualizamos el comportamiento de nuestra estimación en función del número de iteraciones.

```python
import random
import matplotlib.pyplot as plt

n = int(input("Numero de puntos que desea "))
print()

def f(n):
    d = 0
    p = 0
    M = []

    for i in range(n):
        x = random.uniform(0, 5)
        y = random.uniform(0, 3)
        if ((x**2)/25 + (y**2)/9) <= 1 :
          d += 1
        p = 4*(d/(i+1))
        M.append(p)

    fig = plt.figure(1, figsize=(10, 6))
    plt.plot(M, color="green", markersize=1)
    plt.title("Grafica")
    plt.xlabel("Puntos")
    plt.ylabel("Estimación de $\pi$")
    plt.grid()
    plt.show()

    print ()
    print("El valor estimado de pi es" , 4*(d/n))

f(n)
```
